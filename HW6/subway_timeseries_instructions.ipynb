# Task 0: Data load and cleaning

```python
# Import packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%pylab inline
```

```python
# Load data
mta = np.load("MTA_Fare.npy")

print mta.shape #600 stations, 23 ticket types, 194 weeks
print type(mta)
#print data[0][0][0]
```

```python
# Clean data, replace -1s with nans
mta = mta.astype(float)
mta[mta==-1] = np.nan
```

ride types list

```python
ridetype = np.array(['10t', '14d', '1d', '2t', '30d', '7d', 'ada', 'afas', 'exp','ez',
            'ff','mc','mr','month','rfm','rr','sen','spec','tcmc','t',
            'umlim','xbus','stud'])
print len(ridetype)
```

```python
summed = np.nansum(mta, axis=0)
print summed.shape
```

# Task 1: Event detection: Identify the most prominent event. There is a very significant drop (>3-sigma) in all time series. Identify it and figure out what it is due to.

```python
# Create Pandas df
data = pd.DataFrame({})

weeks = pd.date_range(start='2010-05-21', periods=194, freq='W-FRI')
#print weeks
#print type(weeks)

for i in range(0, len(ridetype)):
    data[ridetype[i]] = pd.Series(summed[i], index=weeks)

print data.columns
print data.head()
```

```python
data['Total'] = data.sum(axis=1)
print data.head()
print data.shape
```

```python
plt.figure(figsize=(12,10))
plt.plot(data.index, data.Total)
plt.title('Weekly MTA Rides Over Time', size = 24)
plt.xlabel('Date (week)', size = 14)
plt.ylabel('Total number of rides', size = 14)
plt.show()
# We observer a drop for Hurricane Sandy in late Oct. 2012
```

```python
rolling = pd.rolling_mean(data, 5)

plt.figure(figsize=(12,10))

plt.plot(data.index, data.Total, color = 'blue', alpha = 0.25)
plt.plot(rolling.index, rolling.Total, color = 'green')
plt.title('5 Week Rolling Avg. of Weekly MTA Rides Over Time', size = 24)
plt.xlabel('Date (week)', size = 14)
plt.ylabel('Number of rides', size = 14)
plt.show()
# We observer a drop for Hurricane Sandy in late Oct. 2012
```

```python
plt.figure(figsize=(12,10))

plt.plot(data.index, data.Total - rolling.Total, color = 'blue')
plt.title('Deviation from rolling average', size = 24)
plt.xlabel('Date (week)', size = 14)
plt.ylabel('Number of rides', size = 14)
plt.show()
```

# Task 2: Some of the time series are stationary, some have a complex structure, some show a downward trend: Identify the ridership types that have steadily increased in popularity, and that have steadily decreased. by how much? (e.g what is the ratio of usage in the first 10 and last 10 weeks)

```python
rolling10 = pd.rolling_mean(data, 10)

plt.figure(figsize=(12,10))
for r in ridetype:
    plt.plot(rolling10.index, rolling10[r]/mean(rolling10[r]), label = r)
plt.title('5 Week Rolling Avg. of Weekly MTA Rides Over Time', size = 24)
plt.xlabel('Date (week)', size = 14)
plt.ylabel('Number of rides', size = 14)
plt.legend()
plt.show()
# We observer a drop for Hurricane Sandy in late Oct. 2012
```

```python
# Calculate ratio of fluxes
week10 = rolling10.index[10]
last = rolling10.index[-1]

for r in ridetype:
    print r, round(rolling10[r][last]/rolling10[r][week10],2)
    
# highest increases are present for:
# mail and ride (increased by factor or 25670)
# railroad (" 2.58)
# senior (" 2.88)

# most significant descreases for:
# ada farecard access system (0.0)
# express (0.0)
# ez pay (0.0)
# monthly (0.57)
# reduced fare media (0.62)
# unlimited (0.79)
```

# Task 3: Several stations show a prominent annual periodicity. Identify the 4 stations (indentify them by the index of their location in the data cube) that show the most prominent periodic trend on an annual period (52 weeks). (Can you figure out what the periodic peak in rides is due to?)

```python
stations = np.nansum(mta, axis=1)
print stations.shape

# Create Pandas df
data2 = pd.DataFrame({})

weeks = pd.date_range(start='2010-05-21', periods=194, freq='W-FRI')
#print weeks
#print type(weeks)

for i in range(0, 600):
    data2['Station' + str(i)] = pd.Series(stations[i], index=weeks)

print data2.columns
print data2.head()
```

```python
N = len(data2)
print N

f = np.abs(np.fft.rfft(data2.Station0))
fig=plt.figure(figsize=(15,5))
ax=fig.add_subplot(111)
ax.plot(np.fft.rfftfreq(N, 1.0)[1:], (f[1:]), 'o', ms=20)
ax.plot(np.fft.rfftfreq(N, 1.0)[1:], (f[1:]), '-')
ax.set_xticklabels([ "%s"%(1/f) for f in ax.get_xticks()], fontsize=20)
plt.xlabel("period (week)", fontsize=20)
```

```python
print 1./52
# 5th element of FFT is the closest

fft52 = []
for i in range(0,600):
   fft52.append(np.abs(np.fft.rfft(data2['Station' + str(i)]))[5])
#print fft52

period52 = pd.DataFrame({})
period52['score'] = fft52

print period52.sort(['score'], ascending=[0])[0:4]
# Stations 0, 19, 150, 83 show the stations exhibiting strongest annual period (52 weeks)
```
