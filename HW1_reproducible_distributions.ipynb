# I want you to create a <i>reproducible research</i> code
that means we want the reader to be able to reproduce every step of our analysis. when stochastic processes and random variables are involved that gets tricky, but since computers are unable to rpoduce truly random numbers, we can actually reproduce every series of "pseudo random" number generated by a computer as long as we know what seed is used. <i>The first seed will determin entirely the distribution that is generated, and the seeds of the following distriburions generated by the code!</i>

```python
import os
import sys
import numpy
import matplotlib.pyplot as pl
%matplotlib inline
```

#THIS IS THE MOST IMPORTANT PIECE OF THIS CODE! 
##we must give numpy as seed for this code to be reproducible
http://docs.scipy.org/doc/numpy/reference/generated/numpy.random.seed.html

```python
from numpy.random import RandomState
rand1 = RandomState(616)
```

##use this function to generate 100 2-dimensional random datapoints in a gaussian distribution in 2 dimensions

http://docs.scipy.org/doc/numpy/reference/generated/numpy.random.randn.html

```python
a = rand1.randn(100, 2)
```

## use the same function to generate 50 more random objects, but this time change the mean
to do that you multiply the distribution by a different sigma, and add your new mean - remember that these are 2D objects, so yor mean must be a 2D vector (a list, an array or a tuple)! 


```python
b = 5 + rand1.randn(50, 2)
```

##now plot them with this function: 

https://scipy-lectures.github.io/intro/matplotlib/matplotlib.html#scatter-plots

##if you print the array it will be a 2x100: 100 2D objects. although that is what we wanted we cannot plot it easily this way: there are many ways to deal with it. the goal is to pass the first dimension as the disrt argument to scatter, and the second dimension as the second argument. i am gonna do it by transposing the array (switching the axes) 



```python
pl.clf()
pl.scatter(a[:,0], a[:,1])
pl.show()
```


##first plot them as a single distribution, then change the color for the 2 samples

remember to give your labels some names and to make a legend! in the legend tell me what the mean and standard deviation of each sample is

```python
pl.figure(figsize=(12, 6))

# First, plotted as a single distribution
pl.subplot(1, 2, 1)
pl.scatter(a[:,0], a[:,1])
pl.scatter(b[:,0], b[:,1])
pl.title('Single distribution')

# Next, plotted with different colors
pl.subplot(1,2,2)
orig = pl.scatter(a[:,0], a[:,1], c='blue')
new = pl.scatter(b[:,0], b[:,1], c='violet')
pl.title('Separate distributions')
pl.legend((orig,new), ('Normal distribution with mean 0, sd 1 (100 random points)', 'Normal distribution with mean 5, sd 1 (50 random points)'), scatterpoints = 1, loc='lower center', fontsize = 8)
pl.show()
```
