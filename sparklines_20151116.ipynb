```python
import os
import json
import numpy as np
import matplotlib.pyplot as plt

s = json.load( open(os.getenv ('HOME')+"/fbb_matplotlibrc.json") )
plt.rcParams.update(s)

%pylab inline
```

```python
fig = plt.figure(figsize=(10,12))
ax = []

for i in range(12):
    ax.append(fig.add_subplot(6,2,i+1))
    data = np.random.randn(100)*np.cos(np.arange(100)/np.pi)*np.random.rand()
    ax[i].plot(data, 'k', alpha=0.3)
    plt.xlim(0,130)
    ax[i].axis('off')
    ax[i].plot(np.where(data==max(data))[0], max(data), 'o', color = 'green')
    ax[i].plot(np.where(data==min(data))[0], min(data), 'o', color = 'red')
    ax[i].text(0.8, 0.5,'%.2f'%(max(data)), fontsize=12, transform = ax[i].transAxes, color = 'green', alpha = 0.7)
    ax[i].text(0.95, 0.5,'%.2f'%(min(data)), fontsize=12, transform = ax[i].transAxes, color = 'red', alpha = 0.7)

fig.suptitle('Sparklines', size=24)
```
